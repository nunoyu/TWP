```python
holdings = extra.getSpyHoldings('data')  #提供保存xls文件的目录文件夹
health_cares = holdings[holdings['sector']=='Health Care']
symbols = health_cares['symbol'].tolist()
```

```python
quote = yf.getQuote(symbols)
quote['PE'].hist(bins=100)
xlim([0,60])
```

```python
hist_data = yf.HistData() #创建一个数据对象
hist_data.downloadData(symbols) #多股票数据同时抓取
```

```python
c = hist_data.getDataFrame('close') #仅提取收盘价格成为dataframe
```

```python
figsize(16,8)
c.plot(grid=True, legend=False)
```
