# Leveraged ETF backtest YTD#
Many people think that leveraged etfs in the long term underperform their benchmarks. This is true for choppy markets, but not in the case of trending conditions, either up or down. Leverage only has effect on the most likely outcome, not on the expected outcome. For more background please read this post
2013 has been a very good year for stocks, which trended up for most of the year. Let's see what would happen if we shorted some of the leveraged etfs a year ago and hedged them with their benchmark.
An overview of leveraged etfs can be found here

```python
# 建立一个多行的字符串对象
txt = """
SPY 2 SSO -1 
SPY -2 SDS -1
QQQ 2 QLD -1
QQQ -2 QID -1
IYF -2 SKF -1
"""
s = StringIO(txt) # 在内存中建立对象储存txt
pairs = [] # list of all pairs
symbols = set() # 用set来储存股票名称。set的特点是不会重复记录同名元素。
for line in s.readlines():
    fields = line.split() # splits line into four fields 
    if not fields: continue # skip lines with no fields 
    
    symbols.add(fields[0]) # add first symbol
    symbols.add(fields[2]) # add second symbol
    pairs.append(pd.Series({fields[0]:int(fields[1]), fields[2]:int(fields[3]) })) # create a pair Series and add it
                                                                                   # to the 'pairs' list
print('Symbols: ', symbols)

#---get yahoo data
hData = yahoo.HistData()
hData.downloadData(symbols)
```

```python
# 对一对进行模拟
pair = pairs[0]
print(pair)

price = hData.getDataFrame('close')[pair.index].tail(250)

price = price.dropna() # remove missing rows

price.plot()
ylabel('price [$]')
# 将价格标准化。例如原来是200到300区间，现在变成初始基期为100.
ret = price.pct_change().dropna()
priceNorm = (ret+1).cumprod()
priceNorm.columns = [c+'_norm' for c in price.columns] # add '_norm' to column names. 
(100*priceNorm).plot(ax=gca()) # plot to existing chart
title('Price & normalized price')
ylabel('price')

#savefig('output/leveraged_pair.png')

figure()
shares = (pair*100)/price.iloc[0,:] # pair里面确定了2和-1的long&short比例。计算能买多少股。 

port = (price*shares).sum(1)-100 #横向long的减去short的计算实际持仓金额。减去100不明白为什么
port.plot()
title('Pnl for ' + str((100*pair).to_dict()).strip('{}')) 
ylabel('pnl [$]')
#很多人认为sso应该underperform其track的benchmark，即spy。
#实测看来，并不一定。如果完美track，对冲之后应该是0.这里出现了负值。说明outperform。
```
