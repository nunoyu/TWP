```python
returns = randn(1000, 5000)
price = 100*(returns/100+1).cumprod(axis=0)
figsize(10, 8)
h = plot(price[:,:500])
```

```python
finalprice = price[-1,:]
finalprice 
```

```python
figsize(6, 4)
return3 = returns * 3
price3 = 100 * (return3 / 100 + 1).cumprod(axis=0)
finalprice3 = price3[-1,:] #左侧只用-1，产生 一维数组，如果带冒号-1:就是多维数组，无法画hist
h2 = hist(finalprice3,500)
h1 = hist(finalprice, 100)
xlim([0, 400]) #设定边界
print('Mean:', finalprice.mean())
print('Standard Deviation:', finalprice.std())
```
